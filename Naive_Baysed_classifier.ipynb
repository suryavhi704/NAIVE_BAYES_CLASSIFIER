## **IMPORTING REQUIRED FILES:-**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score, roc_auc_score
```

image.png

## **READING THE DATASET:-**

```python
df=pd.read_csv('emails.csv')
df.head(10)
```

## **DATA PERPROCESSING:-**

```python
#checking for null values:-
df.isnull().sum()
```

```python
## checking for duplicate values:-
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
#cheking for if the data is imbalance or not.
df['spam'].value_counts()
```

## **THE DATA IS IMBALANCE......**

## **VECTORIZATION:-**

image.png

```python
vn=CountVectorizer()
vectorized_data=vn.fit_transform(df['text'])
```

```python
vectorized_data.shape
```

```python
vn.get_feature_names_out()[6000:6100]
```

## **ML PROCESSES:-**



```python
x=vectorized_data
y=df['spam']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y)
```

## THERE ARE THREE TYPES OF NAIVEBAYES CLASSIFIER:-
 ## 1.GAUSSIAN NB.
 ## 2.BERNOULLI NB.
 ## 3.MULTINOMINAL NB.(ADVANCED VERSION OF BERNOULLI).

image.png

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model=MultinomialNB()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
roc_auc_score(y_test,y_pred)
```

## **CHECKING FOR THE PRACTICAL IMPLEMENTATION:-**

```python
emails = ['Congratulations, you have a pre-approved loan of 5000 US Dollars. No documentation required.',
          'Can we have a quick meeting tomorrow at 5 PM ?']
```

```python
vectorized_email=vn.transform(emails)
```

```python
model.predict(vectorized_email)
```
